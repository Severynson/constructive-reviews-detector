## Add a feature column for model training

```python
df["review"] = (
    "TITLE: " + df["title"].str.replace("\n"," ") +
    " | TEXT: " + df["text"].str.replace("\n"," ")
)
```

## Balance classes
#### Find smallest class size & downsample both to the same size

```python
df0 = df[df["constructive"] == 0]
df1 = df[df["constructive"] == 1]

n = min(len(df0), len(df1))

df0_bal = df0.sample(n, random_state=42)
df1_bal = df1.sample(n, random_state=42)

df = pd.concat([df0_bal, df1_bal]).sample(frac=1, random_state=42)
```

## Saving the clean dataset

```python
df[["review", "constructive"]].to_csv(
    "../data/reviews.tsv",
    sep="\t",
    index=False
)
```
